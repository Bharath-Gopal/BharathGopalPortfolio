Obesity has been an issue in the United States over the last few years.
Our project aims at answering the question: 'Does Income have an effect on the prevanlence of Obesity in the United States?'

Our dataset consists of data regarding the prevalence of obesity in men and women for the years 2009 and 2011 as well as the average state incomes for the same years. 

Our datasets come from 3 main sources.
1. The state-wise income for the year 2011, obtained from Wikipedia.
2. The prevalence of obesity in men and women for the years 2009 and 2011 from healthdata.org
3. The state-wise income for the year 2009, obtained from https://nces.ed.gov/programs/digest/d10/tables/dt10_025.asp
